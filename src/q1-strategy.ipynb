# 沪深300ETF 最近5年历史数据

使用 yfinance 获取上交所沪深300ETF（510300.SS）日线数据，并绘制收盘价走势。

```python
import platform

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 设置 matplotlib 中文显示支持
if platform.system() == 'Darwin':
    plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'STHeiti']
elif platform.system() == 'Windows':
    plt.rcParams['font.sans-serif'] = ['SimHei']
else:
    plt.rcParams['font.sans-serif'] = ['Noto Sans CJK SC', 'WenQuanYi Micro Hei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 获取沪深300ETF最近约5年的历史日线数据
df = yf.download(
    '510300.SS',
    start='2021-01-01',
    auto_adjust=True,
    multi_level_index=False,
    progress=False,
)

# 校验后续分析需要的数据结构
df.index = pd.to_datetime(df.index)
assert isinstance(df.index, pd.DatetimeIndex), '索引不是 DatetimeIndex'
assert 'Close' in df.columns, '数据中缺少 Close 列'
assert not df.empty, '未获取到行情数据'

print(df.head())
print(f'\n数据时间范围：{df.index.min().date()} 到 {df.index.max().date()}')
print(f'总行数：{len(df)}')

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], linewidth=1.4, label='收盘价')
plt.title('沪深300ETF (510300) 收盘价走势')
plt.xlabel('日期')
plt.ylabel('收盘价（元）')
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
